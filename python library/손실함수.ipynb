## 에러, 손실, 비용 함수의 모양이 다음과 같다라고 가정, 

$$
    f(x) = x^2 - 4x + 6
$$

- 1변수일 때, 그림을 그리기 제일 쉽고, 
- 손실함수에서 함수의 값을 가장 작게 하는 x를 찾고 싶은겁니다. 

```python
def f(x):
    return x ** 2 - 4 * x + 6
```

```python
x = np.linspace(-5, 5, 101)
y = f(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def grad_fx(x):
    return 2 * x - 4
```

```python
x0 = 0
learning_rate = 0.01
maxIter = 1000
paths = []

for it in range(maxIter):
    paths.append(x0)
    x1 = x0 - learning_rate * grad_fx(x0)
    # print('arg x: {}, slope: {}, f value: {}'.format(x0, grad_fx(x0), f(x0)))
    x0 = x1
```

```python
x = np.linspace(0, 2.5, 1000)
paths = np.array(paths)
plt.plot(x,f(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.plot(paths, f(paths), 'o-')
plt.show()
```

```python
plt.plot(f(paths), 'o-')
plt.grid()
plt.xlabel('x')
plt.ylabel('loss')
plt.show()
```

## non-convex function

$$
    \min_x x sin(x)
$$

- convex하지 않은 함수를 하나 가정
- 손실(에러)함수의 모양이 이런 모양이라면 ... 

```python
x = np.linspace(-3, 10, 501)
f = lambda x:x * np.sin(x)
y = f(x)
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
grad_fx = lambda x: np.sin(x) + x * np.cos(x)
```

```python
x0 = 7
learning_rate = 0.1
maxIter = 10
paths = []

for it in range(maxIter):
    paths.append(x0)
    x1 = x0 - learning_rate * grad_fx(x0)
    # print('arg x: {}, slope: {}, f value: {}'.format(x0, grad_fx(x0), f(x0)))
    x0 = x1
```

```python
paths = np.array(paths)
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.plot(paths, f(paths), 'o-')
plt.show()
```

```python
plt.plot(f(paths))
plt.grid()
plt.xlabel('x')
plt.ylabel('loss')
plt.show()
```

## 이변수일 경우에 그림을 그려 봅시다. 


$$
    f(x,y) = (x-2)^2 + (y-2)^2
$$

- 그림을 그려볼 수 있는 마지막 입니다. 

```python
from visualize import *
```

```python
def f(x, y):
    return (x-2) ** 2 + (y-2) ** 2

def grad_fx(x, y):
    return 2 * x - 4

def grad_fy(x, y):
    return 2 * y - 4
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 25), np.linspace(-5, 5, 25))
z = f(x,y)
```

```python
argmin = np.array([2,2])
argmin = argmin.reshape(-1, 1)
argmin
```

```python
surf(f, x, y, minima=argmin)
```

```python
contour_with_quiver(f, x, y, grad_fx, grad_fy, minima=argmin)
```

```python
x0 = np.array([4, -4])
learning_rate = 0.2
maxIter = 10
paths = [x0]

for it in range(maxIter):
    paths.append(x0)   
    # grad = np.array([grad_fx(x0[0], x0[1]), grad_fy(x0[0], x0[y])])
    grad = np.array([grad_fx(*x0), grad_fy(*x0)])
    x1 = x0 - learning_rate * grad
    # print('arg x: {}, slope: {}, f value: {}'.format(x0, grad, f(*x0)))
    x0 = x1

paths = np.array(paths)
paths = np.array(np.matrix(paths).T)
```

```python
contour_with_path(f, x, y, paths, minima=np.array([[2], [2]]))
```

### 조금 더 연습해 봅시다. 

\begin{equation}
f(x, y) = 3(x-2)^2 + (y-2)^2
\end{equation}

```python
f = lambda x,y: 3*(x-2) ** 2 + (y-2)**2
grad_fx = lambda x,y: 6 * (x-2)
grad_fy = lambda x,y: 2 * (y-2)
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 101), np.linspace(-5, 5, 101))
z = f(x,y)
```

```python
x0 = np.array([-4, -4])
learning_rate = 0.2
maxIter = 10
paths = [x0]

for it in range(maxIter):
    paths.append(x0)   
    # grad = np.array([grad_fx(x0[0], x0[1]), grad_fy(x0[0], x0[y])])
    grad = np.array([grad_fx(*x0), grad_fy(*x0)])
    x1 = x0 - learning_rate * grad
    # print('arg x: {}, slope: {}, f value: {}'.format(x0, grad, f(*x0)))
    x0 = x1

paths = np.array(paths)
paths = np.array(np.matrix(paths).T)
```

```python
argmin = np.array([2,2])
argmin = argmin.reshape(-1, 1)
argmin
```

```python
surf(f, x, y, minima=argmin)
```

```python
# contour_with_quiver(f, x, y, grad_fx, grad_fy, minima=argmin)
contour_with_path(f, x, y, paths, minima=np.array([[2], [2]]))
```
