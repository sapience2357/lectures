## 선형적인 관계로 가정(모수적 추정)
- y = ax + b
- 주어진 데이터를 가장 잘 설명할 수 있는 선 하나를 찾는 것

## error function(오차함수) or cost function

- Squared Error
- Mean Squared Error(MSE)

$$
    \sum_{i = 1}^n | Y_i - (ax_i + b) | ^2
$$

$$
    min_{a,b} \sum_{i = 1}^n | Y_i - (ax_i + b) | ^2
$$

1. 주어진 선에서: ax + b
2. 어떤 기준으로: squared error
3. 가장 좋은: 최소로 하는
4. 선을 찾는 것입니다: a, b

$$
    (a, b) = argmin_{a, b} \sum_{i = 1}^n | Y_i - (ax_i + b) | ^2
$$

## 2차 함수의 최솟값을 구하기 

$$
    f(x) = x^2 + 4x + 6
$$

```python
def f(x):
    return x ** 2 + 4 * x + 6
```

```python
x = np.linspace(-5, 5, 101)
print(x)
```

```python
y = f(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
xidx = np.argmin(y)
argX = x[xidx]
valueF = f(argX)
print('argX:', argX)
print('f min value:', valueF)
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.plot(argX, valueF, 'xr')
plt.show()
```

## 최급하강법(steepest descent) or 경사하강법(Gradient Descent)

$$
    x_{k+1} = x_k - \alpha \nabla f(x_k)
$$
<br>
<br>
$$
    x^{(k+1)} = x^{(k)} - \alpha \nabla f(x^{(k)})
$$

```python
def grad_fx(x):
    return 2 * x + 4
```

```python
x0 = 0
learning_rate = 0.01
maxIter = 100

for it in range(maxIter):
    x1 = x0 - learning_rate * grad_fx(x0)
    print('arg x: {}, slope: {}, f value: {}'.format(x0, grad_fx(x0), f(x0)))
    x0 = x1
```

## 이변수일 경우

    f(x,y) = (x-2)^2 + (y-2)^2

### 미분계수를 구합시다(편미분)

```python
x = sp.symbols('x')
y = sp.symbols('y')
f = (x-2) ** 2 + (y-2) ** 2
print(sp.diff(f, x))
print(sp.diff(f, y))
```

```python
def f(x, y):
    return (x-2) ** 2 + (y-2) ** 2

def grad_fx(x, y):
    return 2 * x - 4

def grad_fy(x, y):
    return 2 * y - 4
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 101), np.linspace(-5, 5, 101))
z = f(x,y)
```

```python
x0 = np.array([-5, -5])
learning_rate = 0.01
maxIter = 1000

for it in range(maxIter):
    # grad = np.array([grad_fx(x0[0], x0[1]), grad_fy(x0[0], x0[y])])
    grad = np.array([grad_fx(*x0), grad_fy(*x0)])
    x1 = x0 - learning_rate * grad
    print('arg x: {}, slope: {}, f value: {}'.format(x0, grad, f(*x0)))
    x0 = x1
```
