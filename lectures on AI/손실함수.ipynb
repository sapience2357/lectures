```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

# 손실함수

$$
    f(x) = x^2 - 4x + 6
$$

- 1변수일 때, 그림을 그리기 제일 쉽고, 
- 손실함수에서 함수의 값을 가장 작게 하는 x를 찾고 싶은겁니다. 

```python
x = sp.Symbol('x')
f = x ** 2 - 4 * x + 6
display(f)
sp.diff(f, x)
```

```python
def f(x):
    return x ** 2 - 4 * x + 6

def grad_f(x):
    return 2 * x - 4
```

```python
x = np.linspace(-5, 5, 101)
y = f(x)
```

```python
plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,1,2)
plt.plot(x, grad_f(x), 'r-')
plt.grid()
plt.show()
```

## optimize(최적화)

```python
x0 = -4
lr = 0.01
maxIter = 1000
x_list = []

for i in range(maxIter):
    x_list.append(x0)
    x1 = x0 - lr * grad_f(x0)
    print('x0: {}, slope: {}, loss: {}'.format(x0, grad_f(x0), f(x0)))
    x0 = x1
```

## x를 추정하는 과정

```python
x_list = np.array(x_list)
plt.plot(x, y)
plt.plot(x_list, f(x_list), 'o-')
plt.show()
```

## 손실함수의 해석

```python
plt.plot(f(x_list), 'o-')
plt.show()
```

# non-convex function

$$
    \min_x x sin(x)
$$

- convex하지 않은 함수를 하나 가정
- 손실(에러)함수의 모양이 이런 모양이라면 ... 

```python
def f(x):
    return x * np.sin(x)

def grad_f(x):
    return np.sin(x) + x * np.cos(x)
```

```python
x = np.linspace(-3, 10, 101)
y = f(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x0 = 7
lr = 0.9
maxIter = 1000
x_list = []

for i in range(maxIter):
    x_list.append(x0)
    x1 = x0 - lr * grad_f(x0)
    x0 = x1
```

```python
x_list = np.array(x_list)
plt.plot(x, y)
plt.plot(x_list, f(x_list), 'o-')
plt.show()
```

```python
plt.plot(f(x_list), 'o-')
plt.show()
```
