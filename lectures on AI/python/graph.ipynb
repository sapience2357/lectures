## pyplot

### 기본 plot 및 옵션

```python
x = np.linspace(1, 10, 10)
y = x + x ** 2 + x ** 3
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o--r', alpha=1, label='first', markeredgecolor='k')
plt.plot(x, x**2, 'x-b', alpha=0.8, label='second', linewidth=2.5, markersize=10, markeredgecolor='k')
plt.title('pyplot의 기본 사용법', {'fontsize': 20})
plt.xlabel('이제는 한글이 안 깨져요', fontsize=20)
plt.ylabel('여기는 Y', fontsize=20)
plt.xlim(-10, 10)
plt.ylim(-1000, 1000)
plt.axvline(0, C='k')
plt.axhline(0, color='k')
plt.annotate('이것은 주석입니다', (-5, 500), fontsize=20)
plt.plot(-5, 500, 'o', markersize=20)
plt.arrow(-5, 500, 0, -400, width=0.2, head_width=0.9,  head_length=100)
plt.grid()
plt.legend()
plt.show()
```

### 그래프 나누어서 그리기(subplot)

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.plot(x, y, 'o--r', alpha=0.8, label='first', markeredgecolor='k')

plt.subplot(1,2,2)
plt.plot(x, x**2, 'x-b', alpha=0.8, label='second', linewidth=2.5, markersize=10, markeredgecolor='k')
plt.show()
```

### 그래프 나누어서 그리기(subplots)
