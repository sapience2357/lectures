# matplotlib를 이용한 그래프 표현하기 

```python
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')

import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.font_manager
```

## 더미 데이터 생성

```python
x = np.linspace(0, 10, 11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

## 그래프 그리기
- plot함수(선, 그래프), scatter(점), hist(히스토그램) 등을 사용
    - 기본적으로 x,y에 해당하는 값이 필요

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

## 그래프에 주석 추가

```python
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('title')
plt.grid()
plt.plot(x,y)
```

```python
plt.plot(x, x, c='r', ls='--', marker='x', label='$x$')
plt.plot(x, x ** 2, c='g', ls='-', marker='o', label='$x^2$')
plt.plot(x, x ** 3, c='b', ls='-.', marker='s', label='$x^3$')
plt.xticks(x, ['first', 'second', 'third', 'forth', 'fifth'])
plt.yticks([0, 200, 400, 600])

plt.title('여러가지 그래프')
plt.xlabel('엑스축')
plt.ylabel('와이축')
plt.legend(loc=2)
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x, x, c='r', ls='--', marker='x', label='$x$')
plt.plot(x, x ** 2, c='g', ls='-', marker='o', label='$x^2$')
plt.plot(x, x ** 3, c='b', ls='-.', marker='s', label='$x^3$')

plt.title('여러가지 그래프')
plt.xlabel('엑스축')
plt.ylabel('와이축')
plt.legend(loc=2)
```

## 한 화면에 여러 그래프 출력하기 
- subplot

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(y, x, 'g')

plt.subplot(2, 2, 3)
plt.plot(x, x ** 2, 'b')

plt.subplot(2, 2, 4)
plt.plot(x, np.log(x), '#000000')
```

# matplotlib의 또 다른 그림들 

## 막대그래프
- 바차트

```python
x = np.arange(5)
y = [100, 200, 300, 400, 500]

plt.bar(x, y, width=0.9, align='center', color='#123456', edgecolor='#00ffff', linewidth=2)
plt.xlabel('먼지는 모르겠지만 일단 x')
plt.show()
```

```python
x = np.arange(5)
y = [100, 200, 300, 400, 500]

plt.barh(x, y, height=0.9, align='center', color='#123456', edgecolor='#00ffff', linewidth=2, alpha=0.8)
plt.xlabel('먼지는 모르겠지만 일단 x')
plt.show()
```

## 히스토그램
- y가 필요 없기 때문에, 단변량 그림 그리기에 좋습니다. 
- bins로 히스토그램의 바의 갯수를 설정

```python
data = np.random.randint(1, 100, size=200)
print(data)
```

```python
data
```

```python
plt.figure(figsize=(20, 7))
plt.hist(data, bins=20, alpha=0.3)
plt.grid()
plt.show()
```

## Scatter(산점도)

```python
N = 50
x = np.random.randn(N)
y = np.random.randn(N)
area = (20 * np.random.rand(N))**2
color = np.sqrt(area)
plt.scatter(x, y, s=area, c=color, alpha=0.8)
plt.show()
```
